## (1) groupby(['성별'])['title']['rating'].mean() -- 시험문제
### pandas.pivot_table(data=, values=, index= , columns=, aggfunc= ) -- 시험문제
* 주어진 데이터프레임을 특정 열을 기준으로 그룹화하기
* 다양한 집계 함수를 사용하여 요약 통계를 계산
* 주요 파라미터
  * data : 피벗 테이블을 생성할 데이터프레임
  * values : 집계할 데이터 열
  * index: 피벗 테이블의 행으로 사용할 열
  * columns: 피벗 테이블의 열로 사용할 열
  * aggfunc: 집계함수 목록
  * fill_value: 결측값을 대체할 값

```python
# (2) 성별에 따른 각 영화의 평균 평점을 집계하기
mean_ratings = pd.pivot_table(data = df, values = "rating", index="title",
                                columns="gender", aggfunc="mean")
mean_ratings.head()
```

```python
# (3) 영화 제목별 건수 집계하기
# 데이터를 영화 제목으로 그룹화하여 size() 함수를 사용
# 동일한 영화 제목을 가지는 행들을 계산하여 반환
ratings_by_title = df.groupby("title").size()
ratings_by_title.head()
```

```python
# (4) 250건 이상의 평점 정보가 있는 영화 제목 추출
active_titles = ratings_by_title.index[ratings_by_title >= 250]
active_titles
```

```python
# (5) 성별에 따른 250건 이상의 평점 정보가 있는 영화에 대한 색인 추출하기
mean_ratings = mean_ratings.loc[active_titles]
mean_ratings
```

### [문제2] 장르와 연령별로 그룹하여 평균 평점을 구하기 -- 시험 범위임

```python
# 여러 장르에 속하는 영화일 경우 장르를 개별적으로 분리하기
movies["genre"] = movies.pop("genres").str.split("|") # 기말 시험에 안나옴
movies.head(4)
```

```python
# (2) 복수 장르에 속하는 영화을 각 장르별로 나누기
movies_exploded = movies.explode("genre")
movies_exploded[:10]
```

```python
# (3) 세 테이블을 모두 병합하여 그룹화하기
ratings_with_genre = pd.merge(pd.merge(movies_exploded, ratings), users)
ratings_with_genre.head(20)
```

```python
ratings_with_genre.shape
```

```python
# (3) 장르와 연령별로 그룹하여 평균 평점을 구하기
# ratings_with_genre.iloc[0]
# (1)
genre_ratings_1 = pd.pivot_table(data = ratings_with_genre, values = "rating",
                                 index = "genre", columns = "age", aggfunc = "mean")
genre_ratings_1[:5]
# (2)
genre_ratings = (ratings_with_genre.groupby(["genre", "age"])
                 ["rating"].mean().unstack("age")
                 )
genre_ratings[:10]
```

## [데이터사이언스]
* 선형회귀 분석
  * 각 장르별로 연령대와 평점 간의 관계를 분석하는 선형 회귀 모델을 구축하고, 그 성능을 비교
  * age를 독립 변수로, 각 장르별 평점을 종속 변수로 사용
  * LinearRegression

```python
# 1. 필요한 라이브러리 임포트
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

# 2. 데이터프레임 생성
df = genre_ratings

# 3. 독립변수 X : age 독립 변수로 설정 2차원 배열-
X = np.array([1, 18, 25, 35, 45, 50, 56]).reshape(-1, 1)

# 4. 각 장르에 대해 선형 회귀 모델을 훈련하고 평가, 그 결과를 리스트에 저장하고 시각화
_, axs = plt.subplots(nrows=9, ncols=2, figsize=(15, 45))
axs = axs.flatten() # 1차원 벡터로 평탄화
results = []

for idx, genre in enumerate(df.index):
    y = df.loc[genre].values # 종속 변수 y: 장르의 평균평점을 추출
    # linearRegression 모델을 사용하려 각 장르에 대해 선형 회귀 분석을 수행
    model = LinearRegression()
    model.fit(X, y)
    y_pred = model.predict(X)
    
    # 모델의 성능을 MSE와 R^2 스코어로 측정
    mse = mean_squared_error(y, y_pred)
    r2 = r2_score(y, y_pred)
    # results 리스트에 성능 지표값 저장
    results.append({
        'genre': genre,
        'mse': mse,
        'r2': r2
    })

    # 각 장르에 대해 실제 값과 예측 값을 시각화
    axs[idx].scatter(X, y, color='blue') # 산포도
    axs[idx].plot(X, model.predict(X), color='red') # 회귀선
    axs[idx].set_title(f'{genre} - Actual vs Predicted')
    axs[idx].set_xlabel('Age')
    axs[idx].set_ylabel('Rating')

plt.tight_layout()
plt.show()

# 모든 장르의 결과를 데이터프레임으로 저장
results_df = pd.DataFrame(results)
print(results_df)

# 모든 장르의 MSE와 R^2 스코어를 막대그래프로 시각화 - 데이터프레임의 시각화
results_df.set_index('genre', inplace=True)
results_df[['mse', 'r2']].plot(kind='bar', subplots=True, layout=(2, 1), 
                               figsize=(10, 10), title="Model Performance by Genre")
plt.show()

# 상관관계를 가장 잘 적합한 장르 찾기- R2 스코어가 가장 높은 장르를 찾아 출력
best_genre = results_df['r2'].idxmax()
print(f'연령과 영화 평점관의 상관', best_genre)
```
